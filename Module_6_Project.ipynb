```python
import networkx as nx
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('sample_data/MADOFF.csv', index_col=0, header=0)
df.head()
```

First, I start by importing the Madoff Scandal data as a Pandas dataframe.

```python
df = df.fillna(0)
df.head()
```

After reviewing the dataframe output, I filled the N/A values with 0, so it will be able to read as a graph.

```python
G = nx.from_pandas_adjacency(df)
```

```python
plt.figure(figsize=(20,10))
nx.spring_layout(G)
nx.draw_networkx(G, font_size=10)
```

I read the data from the CSV file into a NetworkX graph and adjusted the layout to make it more readable. I noticed a loop, which indicates that Bank Medici was funneling money to themselves.  

```python
bc_dict = nx.betweenness_centrality(G)
bc_top5 = dict(sorted(bc_dict.items(), key=lambda x:x[1], reverse=True)[:5])
for key, value in bc_top5.items():
  print(key,':',value)
```

```python
dc_dict = nx.degree_centrality(G)
dc_top5 = dict(sorted(dc_dict.items(), key=lambda x:x[1], reverse=True)[:5])
for key, value in dc_top5.items():
  print(key,':',value)
```

I wanted to see how much exposure each firm had in the scandal, and because there were so many, I just wanted to show the top 5. As one would expect, Bernard Madoff Investment was by far the most involved, with Bank Medici a close second in its involvement.

```python
G_rm = G
G_rm.remove_node('Bank_Medici')
```

When considering the question of the module (which node to remove and the resulting effect), I considered removing the most involved, Madoff Investment, but this felt more like removing the entirety of the network and causing collapse. So instead I want to see what the overall effect to the scandal would be if I instead removed a heavily involved feeder firm, so I removed Bank Medici.

```python
new_bc = nx.betweenness_centrality(G_rm)
new_top5_bc = dict(sorted(new_bc.items(), key=lambda x:x[1], reverse=True)[:5])
for key, value in new_top5_bc.items():
  print(key,':',value)
```

```python
new_dc = nx.degree_centrality(G_rm)
new_top5_dc = dict(sorted(new_dc.items(), key=lambda x:x[1], reverse=True)[:5])
for key, value in new_top5_dc.items():
  print(key,':',value)
```

By taking this approach, I reran the previous functions after removing Bank Medici. We can see here Cohmad Securities falls out of the top 5 for betweenness centrality, which indicates they were closely tied to Bank Medici. Judging by the drop in betweenness and degree centrality for Madoff Investment, the scandal was likely made easier by Bank Medici's involvement. Removing this node exposes how other feeder firms would become more integral to the scandal as well. Looking at this from the viewpoint of someone trying to take down the network, Fairfield Greenwhich would be the next node I would remove. Based on the numbers they would be the next most involved feeder firm in the scandal and with its removal the entire network would continue to collapse. There are pros and cons to starting so close to the top, but sometimes the most effective approach is getting close enough to send a message to the lower feeder firms without wasting ones time with the low hanging fruit. With the ring leader (Madoff Investment) still intact the network will most likely continue on until they are taken down.
